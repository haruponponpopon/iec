```python
# f = open('./fitness_mountain20230108.txt', 'r')
f = open('./experiment_data/20230215164831/fitness.txt', 'r')
datalists1 = f.readlines()
f.close()

f = open('./experiment_data/20230215221349/fitness.txt', 'r')
datalists2 = f.readlines()
f.close()

f = open('./experiment_data/20230216032542/fitness.txt', 'r')
datalists3 = f.readlines()
f.close()

f = open('./experiment_data/20230216083650/fitness.txt', 'r')
datalists4 = f.readlines()
f.close()

f = open('./experiment_data/20230216134823/fitness.txt', 'r')
datalists5 = f.readlines()
f.close()
```

```python
def num_array(datalists):
    data = []
    for datalist in datalists:
        a = []
        num = 0
        divide = 1.0
        flag = False
        mul = 1
        for d in datalist:
            if '0'<=d and d<='9':
                num = num*10+int(d)
                if flag:
                    divide/=10
            elif d=='.':
                flag=True
            elif d==',':
                a.append(num*divide*mul)
                num=0
                flag=False
                divide = 1.0
                mul=1
            elif d=='-':
                mul=-1
        a.append(num*divide*mul)
        data.append(a)
    return data
data1 = num_array(datalists1)
data2 = num_array(datalists2)
data3 = num_array(datalists3)
data4 = num_array(datalists4)
data5 = num_array(datalists5)
```

```python
# average = []
# minimum = []
# maximum = []
# for ar in data:
#     ttl = 0
#     mi = 1
#     ma = -1
#     for a in ar:
#         ttl += a
#         mi = min(mi,a)
#         ma = max(ma,a)
#     average.append(ttl/len(ar))
#     minimum.append(mi)
#     maximum.append(ma)
data = []
for i in range(100):
    data.append(0)
for i in range(100):
    for a in data1[i]:
        data[i] += a
    for a in data2[i]:
        data[i] += a
    for a in data3[i]:
        data[i] += a
    for a in data4[i]:
        data[i] += a
    for a in data5[i]:
        data[i] += a
    data[i]/=150
print(data)
```

```python
#適合度
import numpy as np
import matplotlib.pyplot as plt
font_size=23

x = np.arange(0, len(data)) #-5から5まで0.1区切りで配列を作る

plt.plot(x,data) # この場合のplot関数の第一引数xは、x軸に対応し、第二引数のyがy軸にあたります。
# plt.plot(x,minimum,label='minimum')
# plt.plot(x,maximum,label='maximum')
# plt.legend() #凡例
plt.title("change in fitness",fontsize=font_size)
plt.xlabel('generation',fontsize=font_size) #X軸ラベル
plt.ylabel('fitness',fontsize=font_size) 
plt.tick_params(labelsize=font_size)
plt.yticks(np.arange(0.4, 1.1, step=0.2))
# plt.fill_between(x, minimum, average, facecolor='orange', alpha=0.5)
# plt.fill_between(x, maximum, average, facecolor='green', alpha=0.5)
plt.savefig("graph.pdf", bbox_inches='tight')
plt.show()
```
