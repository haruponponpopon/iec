```python
# f = open('./fitness_mountain20230108.txt', 'r')
f = open('./fitness_data/fitness_mountain20230111.txt', 'r')
datalists = f.readlines()
f.close()
print(datalists[0])
datalists
```

```python
data = []
for datalist in datalists:
    a = []
    num = 0
    for d in datalist:
        if '0'<=d and d<='9':
            num = num*10+int(d)
        elif num!=0:
            a.append(num)
            num=0
    data.append(a)
print(data)
```

```python
average = []
minimum = []
maximum = []
for ar in data:
    ttl = 0
    mi = 50
    ma = 0
    for a in ar:
        ttl += a
        mi = min(mi,a)
        ma = max(ma,a)
    average.append(ttl/len(data))
    minimum.append(mi)
    maximum.append(ma)
```

```python
#
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib

x = np.arange(0, len(data)) #-550.1
font_size=23
plt.plot(x,average,label='') # plotxxyy
plt.plot(x,minimum,label='')
plt.plot(x,maximum,label='')
plt.legend(fontsize=font_size,bbox_to_anchor=(1.5, 0.7)) #
plt.title("",fontsize=font_size)
plt.xlabel('',fontsize=font_size) #X
plt.ylabel('',fontsize=font_size) 
plt.fill_between(x, minimum, average, facecolor='orange', alpha=0.5)
plt.fill_between(x, maximum, average, facecolor='green', alpha=0.5)
plt.tick_params(labelsize=font_size)
plt.savefig("graph.pdf", bbox_inches='tight')
plt.show()
```
