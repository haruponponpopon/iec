```python
# f = open('./fitness_mountain20230108.txt', 'r')
f = open('./fitness_data/fitness_mountain20230111.txt', 'r')
datalists = f.readlines()
f.close()
print(datalists[0])
datalists
```

```python
data = []
for datalist in datalists:
    a = []
    num = 0
    for d in datalist:
        if '0'<=d and d<='9':
            num = num*10+int(d)
        elif num!=0:
            a.append(num)
            num=0
    data.append(a)
print(data)
```

```python
average = []
minimum = []
maximum = []
for ar in data:
    ttl = 0
    mi = 50
    ma = 0
    for a in ar:
        ttl += a
        mi = min(mi,a)
        ma = max(ma,a)
    average.append(ttl/len(data))
    minimum.append(mi)
    maximum.append(ma)
```

```python
#適合度
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib

x = np.arange(0, len(data)) #-5から5まで0.1区切りで配列を作る
font_size=23
plt.plot(x,average,label='平均') # この場合のplot関数の第一引数xは、x軸に対応し、第二引数のyがy軸にあたります。
plt.plot(x,minimum,label='最小値')
plt.plot(x,maximum,label='最大値')
plt.legend(fontsize=font_size,bbox_to_anchor=(1.5, 0.7)) #凡例
plt.title("世代ごとの適合度の推移",fontsize=font_size)
plt.xlabel('世代',fontsize=font_size) #X軸ラベル
plt.ylabel('適合度',fontsize=font_size) 
plt.fill_between(x, minimum, average, facecolor='orange', alpha=0.5)
plt.fill_between(x, maximum, average, facecolor='green', alpha=0.5)
plt.tick_params(labelsize=font_size)
plt.savefig("graph.pdf", bbox_inches='tight')
plt.show()
```
